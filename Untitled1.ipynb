```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data
concurrent_requests = [10, 100, 500, 1000]
cpu_usage = {
    'REST': [8.5, 9.7, 12.1, 14.2],
    'SOAP': [12.1, 13.6, 15.2, 16.3],
    'GraphQL': [6.7, 7.9, 9.4, 11.0],
    'gRPC': [5.4, 6.2, 7.8, 8.9],
}
memory_usage = {
    'REST': [24.3, 28.5, 35.4, 42.1],
    'SOAP': [40.2, 45.0, 55.3, 60.8],
    'GraphQL': [18.1, 22.3, 28.7, 35.1],
    'gRPC': [16.8, 19.4, 24.1, 29.7],
}
```

```python
sns.set(style="whitegrid")
colors = sns.color_palette("muted", n_colors=4)
```

```python
# Plot CPU Usage
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)
for i, protocol in enumerate(cpu_usage.keys()):
    plt.plot(concurrent_requests, cpu_usage[protocol], marker='o', linestyle='-', color=colors[i], label=protocol)
plt.title("CPU Usage by Protocol", fontsize=14)
plt.xlabel("Concurrent Requests", fontsize=12)
plt.ylabel("CPU Usage (%)", fontsize=12)
plt.xticks(concurrent_requests)
plt.legend(title="Protocol", fontsize=10)
plt.grid(True)
```

```python
# Plot Memory Usage
plt.figure(figsize=(10, 6))
for i, protocol in enumerate(memory_usage.keys()):
    plt.plot(concurrent_requests, memory_usage[protocol], marker='s', linestyle='-', color=colors[i], label=protocol)
plt.title("Memory Usage by Protocol", fontsize=14)
plt.xlabel("Concurrent Requests", fontsize=12)
plt.ylabel("Memory Usage (MB)", fontsize=12)
plt.xticks(concurrent_requests)
plt.legend(title="Protocol", fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Throughput data
concurrent_requests = [10, 100, 500, 1000]
throughput_data = {
    "REST": [96.8, 116.5, 120.9, 115.2],
    "SOAP": [48.2, 43.7, 40.2, 37.4],
    "GraphQL": [5627.4, 5543.3, 5521.8, 5489.2],
    "gRPC": [6823.1, 6702.8, 6640.5, 6589.1],
}
```

```python

protocol_colors = sns.color_palette("muted", len(throughput_data))
fig, ax = plt.subplots(figsize=(12, 7))

for i, (protocol, values) in enumerate(throughput_data.items()):
    ax.plot(
        concurrent_requests,
        values,
        label=protocol,
        color=protocol_colors[i],
        marker="o",
        linewidth=2.5
    )

ax.set_title("Throughput Comparison (Requests per Second)", fontsize=16)
ax.set_xlabel("Concurrent Requests", fontsize=14)
ax.set_ylabel("Throughput (req/s)", fontsize=14)
ax.set_xticks(concurrent_requests)
ax.set_xticklabels(concurrent_requests, fontsize=12)
ax.legend(title="Protocol", fontsize=12)
ax.grid(axis="both", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Latency data
message_sizes = [1, 10, 100]  # Message sizes in KB
operations = ["Create", "Read", "Update", "Delete"]
latency_data = {
    "REST": {
        "Create": [12.4, 20.3, 40.7],
        "Read": [10.2, 18.1, 38.2],
        "Update": [15.3, 21.7, 45.1],
        "Delete": [13.8, 19.4, 42.5],
    },
    "SOAP": {
        "Create": [29.8, 45.2, 110.4],
        "Read": [27.1, 42.5, 105.3],
        "Update": [30.4, 47.8, 115.7],
        "Delete": [28.7, 44.0, 112.1],
    },
    "GraphQL": {
        "Create": [8.5, 13.2, 30.4],
        "Read": [6.8, 11.9, 27.8],
        "Update": [9.1, 14.3, 33.5],
        "Delete": [7.4, 12.8, 29.7],
    },
    "gRPC": {
        "Create": [3.9, 5.8, 13.9],
        "Read": [2.7, 4.3, 12.3],
        "Update": [4.1, 6.2, 15.0],
        "Delete": [3.2, 5.1, 14.5],
    },
}
```

```python
protocol_colors = sns.color_palette("muted", len(latency_data))

fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)
axes = axes.flatten()

for op_idx, operation in enumerate(operations):
    ax = axes[op_idx]
    for i, (protocol, values) in enumerate(latency_data.items()):
        ax.plot(
            message_sizes,
            values[operation],
            label=protocol,
            color=protocol_colors[i],
            marker="o",
            linewidth=2.5,
        )
    ax.set_title(f"{operation} Operation", fontsize=14)
    ax.set_xlabel("Message Size (KB)", fontsize=12)
    ax.set_ylabel("Latency (ms)", fontsize=12)
    ax.set_xticks(message_sizes)
    ax.set_xticklabels([f"{size} KB" for size in message_sizes], fontsize=10)
    ax.grid(axis="both", linestyle="--", alpha=0.7)
    ax.legend(title="Protocol", fontsize=10)

fig.suptitle("Latency Comparison for Different Operations and Message Sizes", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
protocol_colors = sns.color_palette("muted", len(latency_data))
operation_styles = ["-", "--", "-.", ":"]

plt.figure(figsize=(12, 8))

for i, (protocol, operations) in enumerate(latency_data.items()):
    for j, (operation, values) in enumerate(operations.items()):
        plt.plot(
            message_sizes,
            values,
            label=f"{protocol} - {operation}",
            color=protocol_colors[i],
            linestyle=operation_styles[j],
            marker="o",
            linewidth=2.5,
        )

plt.title("Latency Comparison for Different Protocols and Operations", fontsize=16)
plt.xlabel("Message Size (KB)", fontsize=14)
plt.ylabel("Latency (ms)", fontsize=14)
plt.xticks(message_sizes, [f"{size} KB" for size in message_sizes], fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis="both", linestyle="--", alpha=0.7)
plt.legend(title="Protocol - Operation", fontsize=10, bbox_to_anchor=(1.05, 1), loc="upper left")

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```
